```python
import json
```

```python
# Read the JSON file
with open('recipes.json', 'r') as f:
    data = json.load(f)
```

```python
# Create a new list containing only the "text" field
new_data = []
for item in data:
    new_data.append({"text": '\n'.join(item['Method'])})
```

```python
with open('output.jsonl', 'w') as f:
    for item in data:
        json.dump({"text": '\n'.join(item['Method'])}, f)
        f.write('\n')
```
